```python
#creating a function that converts a list to a string
def listToString(s):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    return (str1.join(s))
```

```python
def get_menus(url):
    myfile = requests.get(url)
    soup = BeautifulSoup(myfile.content)
    
    menu = soup.findAll("div", {"class": "menu-item-header__3xwnFL-n"})
    #descriptions = soup.findAll("p", {"class": "menu-description__2HXkC4oE"})
    #menu = headers + descriptions
    
    for w in range(0, len(menu)):   
        menu[w] = str(menu[w])                     #converting the objects to strings
        menu[w] = re.sub('<[^>]+>', '', menu[w])   #eliminaiting the scripts in between <>
        
    menu = [' '.join(w for w in menu)] 
    
    menu = listToString(menu)
    menu = menu.lower() #lowercasing every word
    
    menu = re.sub('[^a-zA-ZÀ-ÿ.\s]', '', menu) #removing all the numbers and special characters
    menu = menu.replace(".", "")

    return menu

#testing the function
test1 = get_menus("https://www.opentable.com/ristorante-piccolo?avt=eyJ2IjoxLCJtIjowLCJwIjowfQ&corrid=171fa7e1-38e4-4dfd-9d76-4d2536d676bc&p=2&sd=2020-06-18+19%3A00")
```

# Removing English word from the corpus

In this section we are going to create Cooking Vocabulary, a corpus containing the most common food and cooking words in English. Then we are going to create a function called remove_english to remove the English words from the tested menus by filtering the menus with the Cooking Vocabulary corpus and ntlk English words corpus. 

```python
#creating a cooking corpus that will be used to filter the English food words

def cooking_corpus_creator(url):
    myfile = requests.get(url)
    soup = BeautifulSoup(myfile.content)
    
    cooking_words = soup.findAll("div", {"class": "wordlist-item"})
    
    for w in range(0, len(cooking_words)):   
        cooking_words[w] = str(cooking_words[w])                     #converting the objects to strings
        cooking_words[w] = re.sub('<[^>]+>', '', cooking_words[w])
    
    return cooking_words

cooking_corpus_creator("https://www.enchantedlearning.com/wordlist/cooking.shtml")   
```

```python
#creating a corpus called cooking_vocabulary that contains the most commong English words related to food and cooking

food_words = cooking_corpus_creator("https://www.enchantedlearning.com/wordlist/food.shtml") 
food_verbs = cooking_corpus_creator("https://www.enchantedlearning.com/wordlist/cooking.shtml") 
cooking_vocabulary = food_words + food_verbs
cooking_vocabulary
```

```python
#creating two stemmers 

from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("ed$|'s$")
stemmer1 = RegexpStemmer("d$")
```

```python
#checking whether these words that are considered both Italian and English words are in the two corpus
#we want to avoid removing an Italian word that it is also considered an English word
possible_eng_it_words = ["spaghetti", "linguine", "fettuccine", "mozzarella", "ravioli", "gnocchi", "pecorino", "romano",
                        "pizza", "gelato", "tiramisù", "risotto"]
```

```python
for w in possible_eng_it_words:
    if w in words.words():
        print(w, "- this corpus contains this word, be careful")
    else:
        print(w, "- this word is not in this corpus")
```

```python
for w in possible_eng_it_words:
    if w in cooking_vocabulary:
        print(w, "- this corpus contains this word, be careful")
    else:
        print(w, "- this word is not in this corpus")
```

```python
#creating a function remove_english to remove all the English words in the test corpus

def remove_english(text):
    
    text = treebank_tokenizer.tokenize(text) #tokenizing the text
    lemmatized_text = [wordnet_lemmatizer.lemmatize(word) for word in text] #lemmatizing the words
    lemmatized_text = [w for w in lemmatized_text if w not in cooking_vocabulary]  #filtering the words using the
                                                                                   # cooking vocabulary corpus
    
    lemmatized_stemmed_text = []

    #stemming the words
    for w in lemmatized_text:       
        w = stemmer.stem(w)
        w = stemmer1.stem(w)
        lemmatized_stemmed_text.append(w)
        
    #filtering the words using the nltk English words corpus
    tokenized_Italian_text = [w for w in lemmatized_stemmed_text if w not in words.words()]
    Italian_text = ' '.join(tokenized_Italian_text)
    
    Italian_text = re.sub('[^a-zA-ZÀ-ÿ.\s]', '', Italian_text) #removing all the numbers and special characters
    
    return Italian_text

remove_english(test1)
```

```python
#passing the function remove_english to the menus retrieved

OpenTable_Test_Corpus = []

for menu in Raw_Test_Corpus:
    menu = remove_english(menu)
    OpenTable_Test_Corpus.append(menu)
```

# Tabula File Processor

As we did for the background corpus, we are going to retrieve some menus for the Test Corpus using Tabula.

```python
from os import listdir
from os.path import isfile, join
```

```python
#Upload the retrieved menus from the directory
dir_base = r"/Users/Matteo/Desktop/NLP/project/FIRe-Fake-Italian-Restaurants-evaluator/data/Test Corpus Menus"

def read_file(filename):
    input_file_text = open(filename , encoding='utf-8').read()
    return input_file_text

    
def read_directory_files(directory):
    file_texts = []
    files = [f for f in listdir(directory) if isfile(join(directory, f))]
    for f in files:
        file_text = read_file(join(directory, f))
        file_texts.append(file_text)
    return file_texts
    
Tabula_Test_Corpus = read_directory_files(dir_base)
print(Tabula_Test_Corpus)
```

```python
#Exploring the object tabula_test_corpus
print(type(Tabula_Test_Corpus))
print(len(Tabula_Test_Corpus))
```

```python
#creating a function menu_cleaner to clean the menus

def menu_cleaner(menu): 
    menu = menu.replace('"', '')
    menu = re.sub('[^a-zA-ZÀ-ÿ.\s]', '', menu)
    menu = menu.lower()
    
    return menu
```

```python
#cleaning the menus in the corpus 

for m in range(0, len(Tabula_Test_Corpus)):
    Tabula_Test_Corpus[m] = menu_cleaner(Tabula_Test_Corpus[m])

Tabula_Test_Corpus
```

```python
#removing the \ns and replacing thme with commas in order to have each dish name separated with by commas 
for menu in range(len(Tabula_Test_Corpus)):
    Tabula_Test_Corpus[menu] = Tabula_Test_Corpus[menu].replace(' \n', ',')
    Tabula_Test_Corpus[menu] = Tabula_Test_Corpus[menu].replace('\n', ',')
```

```python
Tabula_Test_Corpus
```

```python
#removing English words from the Tabula test corpus
for menu in range(0, len(Tabula_Test_Corpus)):
    Tabula_Test_Corpus[menu] = remove_english(Tabula_Test_Corpus[menu])

Tabula_Test_Corpus
```
